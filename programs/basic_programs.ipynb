**Program to calculate simple interest and compound interest**

```python
# interest.py
p = float(input("Principal (P): "))
r = float(input("Rate (%) per year (r): "))
t = float(input("Time (years) (t): "))

simple_interest = p * r * t / 100
print("Simple Interest:", simple_interest)

# Compound interest (compounded annually):
amount = p * (1 + r/100)**t
compound_interest = amount - p
print("Amount:", amount)
print("Compound Interest:", compound_interest)
```

**Program to check whether a given number is prime.**

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

**Program to count the number of digits in a given integer.**

```python
num = int(input("Enter a number: "))

count = 0

if num == 0:
    count = 1
else:
    while num != 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```

**Program to count vowels, consonants, digits, and special characters.**

```python
s = input("Enter a string: ")

vowels = 0
consonants = 0
digits = 0
special = 0

for ch in s:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        if ch.lower() in "aeiou":
            vowels += 1
        else:
            consonants += 1
    else:
        special += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special Characters:", special)
```

**Program to count the frequency of each character in a string.**

```python
# freq_chars.py
s = input("Enter string: ")
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
for ch, count in freq.items():
    print(f"'{ch}': {count}")
```

**Program to define a function to find the factorial of a number.**

```python
# factorial_func_rec.py
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(int(input("n: "))))
```

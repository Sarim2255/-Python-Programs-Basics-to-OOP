````python
 print(len("HELLO"))
 print(len([1, 2, 3, 4, 5]))
 print(len({1: 'One', 2: 'Two', 3: 'Three'}))
````

Operator overloading in Python is a form of compile-time polymorphism. Compile-time polymorphism, also
 known as static polymorphism, occurs during the compilation phase of the program based on the number or
 types of arguments passed to a function or operators used with objects.

````python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def __add__(self, other_point):
    print("self : ",self.x, self.y)
    print("other_point : ",other_point.x, other_point.y)
    return Point(self.x + other_point.x, self.y + other_point.y)
# Example usage
point1 = Point(1, 2)
point2 = Point(3, 4)
result_point = point1 + point2 # Calls __add__ method at compile time
print("result_point : ",result_point.x, result_point.y)
print(2+3) # 5
````

In this example, the + operator is overloaded using the add method. The decision of which method to call
 (add) is made during the compilation phase based on the operator used.

# ***Polymorphism in Class Methods***

Polymorphism in class methods means that different classes can define methods with the same name but behave differently depending on the object calling them. This allows for flexible and reusable code.

````python
class Cat:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def info(self):
    print(f"I am a cat. My name is {self.name}. I am {self.age} years old.")
  def make_sound(self):
    print("Meow")
class Dog:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def info(self):
    print(f"I am a dog. My name is {self.name}. I am {self.age} years old.")
  def make_sound(self):
    print("Bark")

cat1 = Cat("Kitty", 2.5)
dog1 = Dog("Fluffy", 4)
for animal in (cat1, dog1):
  animal.make_sound()
  animal.info()
# animal.make_sound()
````

 Here, we have created two classes Cat and Dog. They share a similar structure and have the same method
 names info() and make_sound().

 However, notice that we have not created a common superclass or linked the classes together in any way.
 Even then, we can pack these two different objects into a tuple and iterate through it using a common animal
 variable. It is possible due to polymorphism.

 ***Example : Method Overriding***

````python
from math import pi
class Shape:
  def __init__(self, name):
    self.name = name
  def area(self):
    pass
  def fact(self):
    return "I am a two-dimensional shape."
  def __str__(self):
    return self.name
class Square(Shape):
  def __init__(self, length):
    super().__init__("Square")
    self.length = length
  def area(self):
    return self.length**2
  def fact(self):
    return "Squares have each angle equal to 90 degrees."
class Circle(Shape):
  def __init__(self, radius):
    super().__init__("Circle")
    self.radius = radius
  def area(self):
    return pi*self.radius**2

a = Square(4)
b = Circle(7)
print(b)
print(b.fact())
print(a.fact())
print(b.area())
````

 Due to polymorphism, the Python interpreter automatically recognizes that the fact() method for object a(Square class) is overridden. So, it uses the one defined in the child class.

 On the other hand, since the fact() method for object b isn't overridden, it is used from the Parent Shape class.

 Here, we can see that the methods such as str(), which have not been overridden in the child classes, are used from the parent class.

# ***Method Overloading in Python***
By default, Python does not support method overloading, but we can achieve it by modifying our methods.

Given a single function sum (), the number of parameters can be specified by you.

This process of calling the same method in different ways is called method overloading.

````python
class Vehicle:
  def __init__(self, name, color, price):
    self.name = name
    self.color = color
    self.price = price
  def show(self):
    print('Details:', self.name, self.color, self.price)
  def max_speed(self):
    print('Vehicle max speed is 150')
  def change_gear(self):
    print('Vehicle change 6 gear')

# inherit from vehicle class
class Car(Vehicle):
  def max_speed(self):
    print('Car max speed is 240')
  def change_gear(self):
    print('Car change 7 gear')

# Car Object
car = Car('Car x1', 'Red', 20000)
car.show()
# calls methods from Car class
car.max_speed()
car.change_gear()
# Vehicle Object
vehicle = Vehicle('Truck x1', 'white', 75000)
vehicle.show()
# calls method from a Vehicle class
vehicle.max_speed()
vehicle.change_gear()
````

# *1. Compile-time Polymorphism (Method Overloading)*

 When multiple methods have the same name but different parameters (number or type).

 ***Example in Python***

 (Python doesnt support overloading directly, but we can mimic it using default parameters.)

````python
class MOP:
  def add(self, a=None, b=None, c=None):
    if a is not None and b is not None and c is not None:
      return a + b + c
    elif a is not None and b is not None:
      return a + b
    else:
      return a
# Example usage
m = MOP()
print(m.add(5, 10))
# Output: 15
print(m.add(2, 3, 4))
````

````python
class Calculator:
  def add(self, a=None, b=None, c=None):
    if a is not None and b is not None and c is not None:
      return a + b + c
    elif a is not None and b is not None:
      return a + b
    elif a is not None:
      return a
    else:
      return 0
calc = Calculator()
print(calc.add(5))
# 5
print(calc.add(5, 10))
# 15
print(calc.add(1, 2, 3))
````

 Same function name add() behaves differently depending on how many numbers you pass.

# *2. Run-time Polymorphism (Method Overriding)*

When a child class provides a specific implementation of a method that is already defined in its parent class.

Run-time polymorphism means the same method name can perform different actions at runtime, depending on which object is calling it.

It is mainly achieved using method overriding  where a child (subclass) provides its own version of a
 method already defined in the parent (superclass).

***Example***

````python
class Animal:
  def sound(self):
    print("Animals make different sounds")
class Dog(Animal):
  def sound(self):
    print("Dog barks")
class Cat(Animal):
  def sound(self):
    print("Cat meows")

# Example usage
a1 = Animal()
a2 = Dog()
a3 = Cat()
a1.sound() # Output: Animals make different sounds
a2.sound() # Output: Dog barks
a3.sound() # Output: Cat meows
````

***Example 1  Animal Sounds***

````python
class Animal:
  def sound(self):
    print("Animals make sounds")
class Dog(Animal):
  def sound(self):
    print("Dog barks")
class Cat(Animal):
  def sound(self):
    print("Cat meows")

# Runtime polymorphism
for animal in [Animal(), Dog(), Cat()]:
  animal.sound()
````

 The same method sound() behaves differently for each object  decided at runtime

# ***What is Method Overriding?***

Method overriding happens when a child class defines a method that has the same name, same parameters, and same return type as a method in the parent class.

In Python, this is how run-time polymorphism is achieved.

***How It Works***

The parent class provides a general method.

The child class redefines (overrides) that method to provide specific behavior.

When you call the method using a child object, Python automatically calls the childs version of the method(not the parents).


***Simple Example - Animal Sounds***

````python
class Animal:
  def sound(self):
    print("Animals make different sounds")
class Dog(Animal): # Dog inherits from Animal
  def sound(self): # Overriding parent method
    print("Dog barks")
# Create objects
a = Animal()
d = Dog()
a.sound()       # Output: Animals make different sounds
d.sound()       # Output: Dog barks
````

 Here, the Dog class overrides the sound() method of the Animal class.

 ---

***Example 2  Using super() to Call Parent Method***

You can still call the parent class version of the method using the super() function.

````python
class Animal:
  def sound(self):
    print("Animal makes sound")
class Cat(Animal):
  def sound(self):
    super().sound()
    print("Cat meows")
c = Cat()
c.sound()
````

 super().sound() lets you use both parent and child behavior.

# ***What is Method Overloading?***

Method Overloading means having multiple methods with the same name but different parameters (number or type).

In many languages like Java or C++, this is done automatically at compile time.

But in Python, true method overloading is not supported directly  because the latest defined method with that name always replaces the previous one.



***So how do we simulate method overloading in Python?***

We can achieve similar behavior using:

-   Default arguments
-  Variable arguments (*args)
-   Type checking (isinstance())

---

***Example 1 - Using Default Arguments***

````python
class Math:
  def add(self, a=0, b=0, c=0):
    return a + b + c
m = Math()
print(m.add())         # 0
print(m.add(5, 10))    # 15
print(m.add(1, 2, 3))  # 6
````

 Here, add() works for 0, 2, or 3 parameters  simulating overloading.

---

***Example 2  Using args(Variable Arguments)***

````python
class Math:
  def add(self, *nums):
    total = 0
    for n in nums:
      total += n
    return total
m = Math()
print(m.add(5))             # 5
print(m.add(5, 10))         # 15
print(m.add(1, 2, 3, 4))    # 10
````

 The *nums parameter allows you to pass any number of arguments  Python packs them into a tuple.

 ---

***Example 3  Type-based Overloading (using isinstance)***

````python
class Display:
  def show(self, data=None):
    if isinstance(data, int):
      print("You entered an integer:", data)
    elif isinstance(data, str):
      print("You entered a string:", data)
    else:
      print("No or unknown data type")
d = Display()
d.show()
d.show(10)
# No or unknown data type
# You entered an integer: 10
d.show("Hello")
# You entered a string: Hello
````

Same method name show() works differently depending on data type.

# ***What is Operator Overloading?***

Operator overloading means giving new meaning to built-in operators (like +, -, *, etc.) when they are used with user-defined objects (like your own classes).

In other words: You can change the behavior of operators for your custom classes.

***Example in Real Life***



```
# 10 + 20 # adds two numbers  30
"A" + "B" # joins two strings  "AB"
```



 Both use +  but behave differently!


 This is operator overloading (Python already overloads it internally).

# ***How to Do Operator Overloading in Python***

Python lets you overload operators by defining special methods (also called magic methods or dunder methods).
image.png

***Example 1 - Overloading + Operator***

````python
class Student:
  def __init__(self, marks):
    self.marks = marks
  def __add__(self, other):
    return self.marks + other.marks
s1 = Student(50)
s2 = Student(60)
print(s1 + s2) # Output: 110
````

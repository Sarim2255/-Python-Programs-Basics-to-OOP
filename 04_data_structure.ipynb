# ***1Ô∏è‚É£ LISTS***

***üîπ Definition:***

A List is an ordered, mutable (changeable) collection in Python that allows duplicate elements and stores multiple data types (integers, strings, floats, etc.).

```
my_list = [10, 20, 30, 40]
```

***üîπ Characteristics:***

‚úÖ Ordered ‚Äì maintains insertion order

‚úÖ Mutable ‚Äì can be changed after creation

‚úÖ Heterogeneous ‚Äì can contain different data types

‚úÖ Allows duplicates

‚úÖ Index-based access

# ***üîπ CRUD Operations:***

---

| Operation  | Description                | Example                                     |
| ---------- | -------------------------- | ------------------------------------------- |
| **Create** | Initialize a list          | `fruits = ["apple", "banana", "mango"]`     |
| **Read**   | Access element using index | `fruits[0] ‚Üí "apple"`                       |
| **Update** | Change an element          | `fruits[1] = "grape"`                       |
| **Delete** | Remove element             | `del fruits[2]` or `fruits.remove("apple")` |


***üîπ Common Built-in Methods:***

---

| Method          | Description                            | Example                      |
| --------------- | -------------------------------------- | ---------------------------- |
| `append(x)`     | Adds an item at the end                | `fruits.append("kiwi")`      |
| `insert(i, x)`  | Insert item at index `i`               | `fruits.insert(1, "orange")` |
| `remove(x)`     | Remove first occurrence of item        | `fruits.remove("apple")`     |
| `pop(i)`        | Remove and return element at index `i` | `fruits.pop(1)`              |
| `sort()`        | Sorts list in ascending order          | `numbers.sort()`             |
| `reverse()`     | Reverses order of list                 | `numbers.reverse()`          |
| `count(x)`      | Returns frequency of item              | `fruits.count("apple")`      |
| `extend(list2)` | Joins two lists                        | `list1.extend(list2)`        |
| `clear()`       | Empties the list                       | `fruits.clear()`             |
| `index(x)`      | Finds index of an item                 | `fruits.index("mango")`      |

# ***üü† 2Ô∏è‚É£ TUPLES***

***üîπ Definition:***

A Tuple is an ordered, immutable (unchangeable) collection that can store multiple data types.

```
student = ("Ali", 21, "MCA")
```

***üîπ Characteristics:***

‚úÖ Ordered

‚ùå Immutable (cannot be changed after creation)

‚úÖ Allows duplicates

‚úÖ Heterogeneous

‚úÖ Faster than lists

---
# ***üîπ CRUD Operations:***

---

| Operation  | Description             | Example                     |
| ---------- | ----------------------- | --------------------------- |
| **Create** | Initialize tuple        | `colors = ("red", "green")` |
| **Read**   | Access via index        | `colors[1] ‚Üí "green"`       |
| **Update** | ‚ùå Not possible          |                             |
| **Delete** | Can delete entire tuple | `del colors`                |


---
***üîπ Common Built-in Methods:***

---
| Method     | Description        | Example      |
| ---------- | ------------------ | ------------ |
| `count(x)` | Counts occurrences | `t.count(5)` |
| `index(x)` | Finds index        | `t.index(3)` |
| `len(t)`   | Returns length     | `len(t)`     |
| `max(t)`   | Maximum value      | `max(t)`     |
| `min(t)`   | Minimum value      | `min(t)`     |

# ***üü† 3Ô∏è‚É£ DICTIONARIES***

***üîπ Definition:***

A Dictionary stores data as key-value pairs, where keys must be unique and immutable, while values can change.
```
student = {"name": "Ali", "age": 21, "course": "MCA"}
```

***üîπ Characteristics:***

‚úÖ Unordered (Python 3.6+ keeps insertion order)

‚úÖ Mutable

‚úÖ Key-value pairs

‚úÖ Keys are unique

---
# ***üîπ CRUD Operations:***

---
| Operation  | Description       | Example              |
| ---------- | ----------------- | -------------------- |
| **Create** | Define dictionary | `d = {"a":1, "b":2}` |
| **Read**   | Access via key    | `d["a"]`             |
| **Update** | Change value      | `d["b"] = 3`         |
| **Delete** | Remove item       | `del d["a"]`         |


---
***üîπ Common Built-in Methods:***

---
| Method                   | Description             | Example                      |
| ------------------------ | ----------------------- | ---------------------------- |
| `keys()`                 | Returns all keys        | `d.keys()`                   |
| `values()`               | Returns all values      | `d.values()`                 |
| `items()`                | Returns key-value pairs | `d.items()`                  |
| `get(key)`               | Safe value access       | `d.get("a")`                 |
| `update(dict2)`          | Adds multiple pairs     | `d.update({"c": 3})`         |
| `pop(key)`               | Removes by key          | `d.pop("a")`                 |
| `clear()`                | Empties dictionary      | `d.clear()`                  |
| `copy()`                 | Creates shallow copy    | `d.copy()`                   |
| `setdefault(key, value)` | Adds if key missing     | `d.setdefault("d", 4)`       |
| `fromkeys(keys, value)`  | Create dict from keys   | `dict.fromkeys(['a','b'],0)` |

# ***üü† 4Ô∏è‚É£ SETS***

***üîπ Definition:***

A Set is an unordered, mutable collection that does not allow duplicates.
```
numbers = {1, 2, 3, 4}
```

***üîπ Characteristics:***

- ‚ùå Unordered (no indexing)

- ‚úÖ Mutable

- ‚ùå No duplicates

- ‚úÖ Supports mathematical set operations (union, intersection, etc.)

---
# ***üîπ CRUD Operations:***

---

| Operation  | Description     | Example       |
| ---------- | --------------- | ------------- |
| **Create** | Define set      | `s = {1,2,3}` |
| **Read**   | Iterate through | `for i in s:` |
| **Update** | Add element     | `s.add(4)`    |
| **Delete** | Remove element  | `s.remove(2)` |


---
***üîπ Common Built-in Methods:***

---

| Method                    | Description         | Example                     |
| ------------------------- | ------------------- | --------------------------- |
| `add(x)`                  | Adds element        | `s.add(5)`                  |
| `remove(x)`               | Removes element     | `s.remove(3)`               |
| `discard(x)`              | Removes safely      | `s.discard(10)`             |
| `clear()`                 | Empties set         | `s.clear()`                 |
| `union(t)`                | Combines sets       | `s.union(t)`                |
| `intersection(t)`         | Common elements     | `s.intersection(t)`         |
| `difference(t)`           | Unique to first set | `s.difference(t)`           |
| `symmetric_difference(t)` | Elements not common | `s.symmetric_difference(t)` |
| `issubset(t)`             | Checks subset       | `s.issubset(t)`             |
| `issuperset(t)`           | Checks superset     | `s.issuperset(t)`           |


````python
# Removing duplicate entries:

data = [1, 2, 2, 3, 3]
unique = set(data)
print(data)
print(unique)
````

````python
# Tracking unique visitors:

visitors = {"user1", "user2", "user1"}
print(visitors)
print(type(visitors))
````

````python
# Finding common students in two courses:

courseA = {"Ali", "Sara"}
courseB = {"Sara", "Zain"}
common = courseA.intersection(courseB)
print(common)
````

````python
# Tag management in blogs:

tags = {"Python", "AI", "ML"}
tags.add("Coding")
print(tags)
````

````python
# Finding differences between two datasets:

old = {1,2,3}
new = {2,3,4}
removed = old.difference(new)
print(removed)
````

````python
# Detecting duplicates in file uploads:

files = ["a.jpg", "b.jpg", "a.jpg"]
unique_files = set(files)
print(files)
print(unique_files)
````

````python
# Managing access roles:

roles = {"read", "write"}
roles.add("delete")
print(roles)
````

````python
# Finding students missing from class:

all_students = {"Ali", "Sara", "Zain"}
present = {"Ali", "Sara"}
absent = all_students.difference(present)
print(absent)
print(present)
````

````python
# Unique product categories:

categories = {"Clothing", "Electronics", "Clothing"}
print(categories)
````

````python
# Keyword extraction from text:

words = set("python is fun and python is powerful".split())
print(words)
````

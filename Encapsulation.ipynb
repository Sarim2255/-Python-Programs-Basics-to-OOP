# ***Encapsulation***

Encapsulation is one of the main principles of Object Oriented Programmin(OOP).
It means hiding the internal details (data) of an object and only exposing what's necessary through methods in Python, encapsulation is done using access modifiers:

**Public:**   

          accessible from anywhere.

**Protected(with one underscore_):**

          meant to be used wothin the class or subclasses.

**Private(with two underscores__):**

          not directly accessible from outside the class.

```python
class Person:
  def __init__(self, name, age):
    self.name = name              #Public attribute
    self._address = "Unknown"     #Protected attribute
    self.__salary = 5000          #Private attribute

  # Public method to access private attribute
  def show_salary(self):
    return self.__salary
  # Public method to modify private attribute
  def set_salary(self,new_salary):
    if new_salary > 0:
      self.__salary = new_salary
    else:
      print("Invalid salary!")


# Create object
p1 = Person("Alice",25)

#Access public attribute
print(p1.name)            #Works fine?

# Access protected attribute (possible, but not recommended).
print(p1._address)        #Should be treated as internal use.

# Access private attribute (won't work directly)
# print(p1.__salary)      #Error: AttributeError.

# Access private attribute using public method.
print(p1.show_salary())

p1.set_salary(700000)
print(p1.show_salary())    #7000
```

# ***Key Points***



*   __salary is private and cannot be accessed directly.
*   You use getter and setter methods to sefely across or modify private data.
*   This keeps your data safe and controlled, which is the main goal of encapsulation.    







**Example 1. Basic Encapsulation with Private Variables**

```python
class Car:
  def __init__(self,brand,speed):
    self.brand = brand
    self.__speed = speed        #Private Attribute
  def get_speed(self):
    return self.__speed
  def set_speed(self,new_speed):
    if new_speed > 0:
      self.__speed = new_speed
    else:
      print(" Speed cannot be negative...")

# Create an object.
car =Car("Totota", 120)
# Access Public attribute
print(car.brand)              #Toyota

# Access private attribute (will cause error)
# print(car.__speed())         # AttributeError.

# Access via methods.
print(car.get_speed())        #120
car.set_speed(150)
print(car.get_speed())        #150
```

# ***Why encapsulation?***

Because __speed is hidden, we can control how it's modified (e.g., prevent negative speeds).

Example 2: BankAccount (Realistic example)

```python
class BankAccount:
  def __init__(self,owner,balance):
    self.owner = owner
    self.__balance = balance        #Private Attribute

  def deposit(self,amount):
    if amount > 0:
      self.__balance += amount
      print(f"Deposited {amount}. New balance: {self.__balance}")
    else:
      print("Deposit amount must be positive")
  def withdraw(self,amount):
    if  0 <amount <= self.__balance:
      self.__balance -= amount
      print(f"Withdrew {amount}. New balance:{self.__balance}")
    else:
      print("Invalid wothdrawl amount!")

  def get_balance(self):
    return self.__balance

#Usage
acc = BankAccount("Alice",1000)
acc.deposit(500)
acc.withdraw(200)
print("Final Balance: ",acc.get_balance())
```

# ***What's encapsulated?***

The __balance attribute-only accessible and modified through controlled methods.

**Example 3: Using @ property (Pythonic Way)**

```python
class Student:
  def __init__(self, name, grade):
    self.name = name
    self.__grade = grade        #Private Attribute

  @property
  def grade(self):
    return self.__grade

  @grade.setter
  def grade(self,value):
    if 0 <= value <= 100:
      self.__grade = value
    else:
      print("Invalid grade. Grade must be between 0 and 100.")

# Example usage
s1 = Student("Alice",85)
print(s1.grade)        #Access using property (acts like a variable)
s1.grade = 95          #Change using setter
print(s1.grade)
s1.grade = 150         #Invalid, rejected
```

*The @property decorator allows safe access and modification of private data in a clean, attribute-like syntax.*






**Example 4: Protected Members**

```python
class Employee:
  def __init__(self, name, salary):
    self.name = name
    self._salary = salary        #Protected (single underscore)
class Manager(Employee):
  def show_salary(self):
    print(f"{self.name}'s salary is: {self._salary}")


m1 = Manager("Emma",8000)
m1.show_salary()        # Access from subclass
print(m1._salary)       # Works, but should be treated as internal
```

Protected memebers are a convention - you can access them, but it's a signal to programmers: "Don't touch this outside of the class or subclasses."

```python
class Student:
  def __init__(self,name,grade):
    self.name = name        # public
    self.__grade = grade    # private

  def get_grade(self):        # public method (getter)
    return self.__grade

  def set_grade(self,grade):  # public method (setter)
    if 0 <= grade <= 100:
      self.__grade = grade
    else:
      print("Invalid grade!")

# Create an instance of the Student class outside the class definition
s1 = Student("Alice",85)

# Access public attribute
print(s1.name)          # works
# print(s1.__grade)     # Error (private)

# Access via getter
print (s1.get_grade())

# Change via setter
s1.set_grade(95)
print(s1.get_grade())   # 95
```

The private attribute is protected using getter/setter methods-a key part of encapsulation.


**Example 3: Protected Variable and Inheritance**

```python
class Parent:
  def __init__(self):
    self._protected_value = 42        # Protected

class Child(Parent):
  def show_value(self):
    print("Accessing protected member from child:",self._protected_value)

c = Child()
c.show_value()               # Allowed in subclass
print(c._protected_value)    # Works, but not recommended outside
```

The single underscore_means "don't touch this outside the class or subclass," but Python doesn't strictly block access.

***Car Class (Public, Protected, and Private)***

```python
class Car:
  def __init__(self,brand,model,price):
    self.brand = brand          # Public attribute
    self._model = model         # Protected attribute
    self.__price = price        # Private attribute

  # Public method
  def show_details(self):
    print(f"Brand: {self.brand}")
    print(f"Model: {self._model}")
    print(f"Price: {self.__price}")

  # Getter (for private attribute)
  def get_price(self):
    return self.__price

  # Setter (to safely modify private attribute)
  def set_price(self, new_price):
    if new_price > 0:
      self.__price = new_price
    else:
      print("Invalid price!")

# Create an object outside the class definition
car = Car("Toyota","Corolla",20000)

# Access public attribute
print("Public: ",car.brand)       # Accessible

# Access protected attribute
print("Protected: ",car._model)    # Accessible, but not recommended

# Access private attribute directly (will fail)
# print(car.__price)                # Error (private)

# Access private attribute via getter
print("Private (via getter):",car.get_price())        # works

# Modify private attribute using setter
car.set_price(2500000)
print("Updated Price:", car.get_price())

# Show all details
print("\nUsing public method:")
car.show_details()
```

***Public Members (Open to Everyone)***

```python
class Dog:
  def __init__(self, name,age):
    self.name = name          # Public
    self.age = age            # Public

dog = Dog("Bruno", 3)
print(dog.name)          # Accessible
print(dog.age)           # Accessible

dog.age = 4              # Can modify
print("Updated age: ", dog.age)
```

## ***Explanation:***

Public members are accessible and changeable from anywhere.

You use them when it's safe for outside code to access and modify the value.

***Private Members (Hidden from Outside)***

```python
class Account:
  def __init__(self,owner,balance):
    self.owner = owner
    self.__balance = balance          # Private attribute

  def show_balance(self):
    print(f"Balance: ${self.__balance}")

acc = Account("Alice",1000)
acc.show_balance()          # Accessed using method
# print(acc.__balance)          # Error! Can't access directly
```

***Explaination:***

__balancw is private, so it's not directly accessible.

We access it safely through the method show_balance().



## *Protected Members (For Internal or Subclass Use)*

```python
class Student:
  def __init__(self, name, grade):
    self.name = name
    self.__grade = grade          # Private

  def get_grade(self):
    return self.__grade
  def set_grade(self,value):
    if 0 <= value <=  100:
      self.__grade = value
    else:
      print("Invalid grade!")

s = Student("Emma",90)
print(s.get_grade())          # 90
s.set_grade(95)               # Valid update
print(s.get_grade())          # 95
s.set_grade(150)              # Invalid
```

***Explaination:***

Getter and setter methods control how private data is accessed or changed.

You can add validation logic (e.g., grade must be 0-100)


***Using @property (Pythonic Way)***

```python
class Employee:
  def __init__(self, name, salary):
    self.name = name
    self.__salary = salary

  @property
  def salary(self):          # Acts like a getter
    return self.__salary
  @salary.setter
  def salary(self, value):   # Acts like a setter
    if value > 0:
      self.__salary = value
    else:
      print("Salary must be positive!")

emp = Employee("Mike",30000)
print(emp.salary)          # Uses @property getter
emp.salary = 40000         # Uses @salary.setter
print(emp.salary)
emp.salary = -500          # Invalid
```

***Explaination:***

@property and @setter make getters and setters liik like normal varialbe access.

This is the cleanest and most Pythonic way to use encapsulation.

```python
# All Three Together (Summary Example)
class Example:
  def __init__(self):
    self.public = "Public Value"
    self._protected = "Protected Value"
    self.__private = "Private Value"
  def show(self):
    print("Inside class:",self.public,self.__protected,self.__private)

obj = Example()
obj.show()

print("\nOutside Class: ")
print(obj.public)          # Public
print(obj._protected)      # Protected (but accessible)
# print(obj.__private)     # Private, can't access
```

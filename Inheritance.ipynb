# **INHERITANCE IN PYTHON**

Inheritance is an Object-Oriented Programming (OOP) concept that allows one class to use the properties and methods of another class.

It helps in code reusability and creating a relationship between classes.

- The class being inherited from is called Parent Class / Base Class

- The class that inherits is called Child Class / Derived Class

```python
# üîπBasic Example of Inheritance
# Parent class
class Animal:
    def sound(self):
        print("Animals make sound")

# Child class inheriting from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating object of Dog class
d = Dog()
d.sound()  # Inherited from Animal
d.bark()   # Own method
```

# ***Types of Inheritance in Python***

---

| Type                     | Description                     | Example       |
| ------------------------ | ------------------------------- | ------------- |
| Single Inheritance       | One child from one parent       | A ‚Üê B         |
| Multiple Inheritance     | One child from multiple parents | A,B ‚Üê C       |
| Multilevel Inheritance   | Chain of inheritance            | A ‚Üê B ‚Üê C     |
| Hierarchical Inheritance | One parent, many children       | A ‚Üê B , A ‚Üê C |
| Hybrid Inheritance       | Combination of above types      | Mix structure |


```python
class BaseClass:
  def base_fun(self):
    print("I am Base Class function")
class ChildClass(BaseClass):
  def child_fun(self):
    print("I am Child class function")


obj = ChildClass()
obj.base_fun()    #Child class can access base function
obj.child_fun()
```

```python
# In this program, you can access the parent class function using the child class object.

class Parent:
  def first(self):
    print('Hello, I am Sam')
class Child(Parent):
  def second(self):
    print("I am Jones")

o = Child()
o.first()
o.second()
```

```python
class ClassA:
  def display(self):
    print('In Class A')
class ClassB(ClassA):
  def display(self):
    ClassA.display(self)
    print('In Class B')

X = ClassB()
X.display()
```
